We now perform the following computations in Python, removing all possible NaN outputs along the way, and plot the market implied volatility smile as needed:

```python
# European vanilla options under the Black-Scholes model
def C_BSv(S0, K, r, sigma, T): # call for European option
    d1 = (np.log(S0/K) + (r + sigma**2/2) * T)/(sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S0 * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

def P_BSv(S0, K, r, sigma, T): # put for European option
    d1 = (np.log(S0/K) + (r + sigma**2/2) * T)/(sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return K * np.exp(-r * T) * norm.cdf(-d2) - S0 * norm.cdf(-d1)

# Market implied volatility under the Black-Scholes model
def impliedVol_BS(S0, K, r, price, T, payoff):
    try:
        if (payoff == "C"):
            impliedVol = brentq(lambda x: C_BSv(S0, K, r, x, T) -
                                price,
                                1e-12,
                                1000.0)
        elif (payoff == "P"):
            impliedVol = brentq(lambda x: price -
                                P_BSv(S0, K, r, x, T),
                                1e-12,
                                1000.0)
        else:
            raise NameError("Payoff type not recognized")
    except Exception:
        impliedVol = np.nan
    return impliedVol

SPX_df1_BS = SPX_df1.copy()
SPX_df2_BS = SPX_df2.copy()
SPX_df3_BS = SPX_df3.copy()

SPY_df1_BS = SPY_df1.copy()
SPY_df2_BS = SPY_df2.copy()
SPY_df3_BS = SPY_df3.copy()

SPX_df1_BS["vol_BS"] = SPX_df1_BS.apply(
    lambda x: impliedVol_BS(SPX_value,
                            x["strike"],
                            r1,
                            x["mid"],
                            T1,
                            x["cp_flag"]), axis = 1)
SPX_df2_BS["vol_BS"] = SPX_df2_BS.apply(
    lambda x: impliedVol_BS(SPX_value,
                            x["strike"],
                            r2,
                            x["mid"],
                            T2,
                            x["cp_flag"]), axis = 1)
SPX_df3_BS["vol_BS"] = SPX_df3_BS.apply(
    lambda x: impliedVol_BS(SPX_value,
                            x["strike"],
                            r3,
                            x["mid"],
                            T3,
                            x["cp_flag"]), axis = 1)

SPY_df1_BS["vol_BS"] = SPY_df1_BS.apply(
    lambda x: impliedVol_BS(SPY_value,
                            x["strike"],
                            r1,
                            x["mid"],
                            T1,
                            x["cp_flag"]), axis = 1)
SPY_df2_BS["vol_BS"] = SPY_df2_BS.apply(
    lambda x: impliedVol_BS(SPY_value,
                            x["strike"],
                            r2,
                            x["mid"],
                            T2,
                            x["cp_flag"]), axis = 1)
SPY_df3_BS["vol_BS"] = SPY_df3_BS.apply(
    lambda x: impliedVol_BS(SPY_value,
                            x["strike"],
                            r3,
                            x["mid"],
                            T3,
                            x["cp_flag"]), axis = 1)

SPX_df1_BS.info()
```

## 1. Displaced-diffusion model
We implement this model in Python using the following code to plot the fitted implied volatility smile under the displaced-diffusion model against the market data:

```python
# European vanilla options under the Black model
def C_Bv(F0, K, r, sigma, T):
    d1 = (np.log(F0/K) + sigma**2/2 * T)/(sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return np.exp(-r * T) * (F0 * norm.cdf(d1) - K * norm.cdf(d2))

def P_Bv(F0, K, r, sigma, T):
    d1 = (np.log(F0/K) + sigma**2/2 * T)/(sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return np.exp(-r * T) * (K * norm.cdf(-d2) - F0 * norm.cdf(-d1))

def calibration_D(x, F0, strike_prices, r, T, vols, payoff):
    err = 0.0
    for i, vol in enumerate(vols):
        err += (vol - impliedVol_D(F0, strike_prices[i], r, x[0], T, x[1], payoff[i]))**2
    return err

# Market implied volatility under the displaced-diffusion model
def impliedVol_D(F0, K, r, sigma, T, beta, payoff):
    try:
        if (payoff == "C"):
            impliedVol = brentq(lambda x: C_Bv(F0/beta, K + (1 - beta)/beta * F0, r, beta * x, T) -
                                sigma,
                                1e-12,
                                1000.0)
        elif (payoff == "P"):
            impliedVol = brentq(lambda x: sigma -
                                P_Bv(F0/beta, K + (1 - beta)/beta * F0, r, beta * x, T),
                                1e-12,
                                1000.0)
        else:
            raise NameError("Payoff type not recognized")
    except Exception:
        impliedVol = np.nan
    return impliedVol

def vol_D(df, F0, strike_prices, r, T, typ, name):
    initialGuess = [0.1, 0.9]
    res = least_squares(lambda x: calibration_D(x,
                                                F0,
                                                df["strike"],
                                                r,
                                                T,
                                                df["implied_vol"],
                                                df["cp_flag"]),
                        initialGuess)
    sigma = res.x[0]
    beta = res.x[1]
    print(f"Calibrated displaced-diffusion model parameters corresponding to asset {typ}'s expiry date {name}: \
    σ = %.3f, β = %.3f" % (sigma, beta))

    # Compute the implied volatility under the displaced-diffusion model using the fitted parameters
    vols_D = []
    for i, K in enumerate(strike_prices):
        vols_D.append(impliedVol_D(F0, K, r, sigma, T, beta, df["cp_flag"][i]))
    return vols_D, sigma, beta

SPX_df1_D = SPX_df1.copy()
SPX_df2_D = SPX_df2.copy()
SPX_df3_D = SPX_df3.copy()

SPY_df1_D = SPY_df1.copy()
SPY_df2_D = SPY_df2.copy()
SPY_df3_D = SPY_df3.copy()

F0_SPX1 = SPX_value * np.exp(r1 * T1)
F0_SPX2 = SPX_value * np.exp(r2 * T2)
F0_SPX3 = SPX_value * np.exp(r3 * T3)
F0_SPY1 = SPY_value * np.exp(r1 * T1)
F0_SPY2 = SPY_value * np.exp(r2 * T2)
F0_SPY3 = SPY_value * np.exp(r3 * T3)

vol_D_x1 = vol_D(SPX_df1_BS, F0_SPX1, strike_prices_x1_BS, r1, T1, "SPX", exdate1)[0]
vol_D_x2 = vol_D(SPX_df2_BS, F0_SPX2, strike_prices_x2_BS, r2, T2, "SPX", exdate2)[0]
vol_D_x3 = vol_D(SPX_df3_BS, F0_SPX3, strike_prices_x3_BS, r3, T3, "SPX", exdate3)[0]
vol_D_y1 = vol_D(SPY_df1_BS, F0_SPY1, strike_prices_y1_BS, r1, T1, "SPY", exdate1)[0]
vol_D_y2 = vol_D(SPY_df2_BS, F0_SPY2, strike_prices_y2_BS, r2, T2, "SPY", exdate2)[0]
vol_D_y3 = vol_D(SPY_df3_BS, F0_SPY3, strike_prices_y3_BS, r3, T3, "SPY", exdate3)[0]
```

## 2. SABR model ($\beta$ = 0.7)
We implement this model in Python using the following code to plot the fitted implied volatility smile under the SABR model against the market data:

```python
beta = 0.7

def calibration_SABR(x, F0, strike_prices, T, vols):
    err = 0.0
    for i, vol in enumerate(vols):
        err += (vol - SABR(F0, strike_prices[i], T, x[0], beta, x[1], x[2]))**2
    return err

def SABR(F0, K, T, alpha, beta, rho, nu):
    X = K
    # if K is ATM forward
    if abs(F0 - K) < 1e-12:
        numer1 = (((1 - beta)**2)/24) * alpha**2/(F0 ** (2 - 2 * beta))
        numer2 = 0.25 * rho * beta * nu * alpha/(F0 ** (1 - beta))
        numer3 = ((2 - 3 * rho**2)/24) * nu**2
        VolAtm = alpha * (1 + (numer1 + numer2 + numer3) * T)/(F0 ** (1 - beta))
        SABR_sigma = VolAtm
    else:
        z = (nu/alpha) * ((F0 * X) ** (0.5 * (1 - beta))) * np.log(F0/X)
        zhi = np.log((((1 - 2 * rho * z + z**2)**0.5) + z - rho)/(1 - rho))
        numer1 = (((1 - beta)**2)/24)*(alpha**2/((F0 * X)**(1 - beta)))
        numer2 = 0.25 * rho * beta * nu * alpha/((F0 * X)**((1 - beta)/2))
        numer3 = ((2 - 3 * rho**2)/24) * nu**2
        numer = alpha * (1 + (numer1 + numer2 + numer3) * T) * z
        denom1 = ((1 - beta)**2/24) * (np.log(F0/X))**2
        denom2 = (((1 - beta)**4)/1920) * ((np.log(F0/X))**4)
        denom = ((F0 * X)**((1 - beta)/2)) * (1 + denom1 + denom2) * zhi
        SABR_sigma = numer/denom
    return SABR_sigma

def vol_SABR(df, F0, strike_prices, r, T, typ, name):
    initialGuess = [0.02, 0.2, 0.1]
    res = least_squares(lambda x: calibration_SABR(x,
                                                   F0,
                                                   df["strike"],
                                                   T,
                                                   df["implied_vol"]),
                        initialGuess)
    alpha = res.x[0]
    rho = res.x[1]
    nu = res.x[2]
    print(f"Calibrated SABR model parameters corresponding to asset {typ}'s expiry date {name}: \
    α = %.3f, β = %.1f, ρ = %.3f, ν = %.3f" % (alpha, beta, rho, nu))

    # Compute the implied volatility under the SABR model using the fitted parameters
    vols_SABR = []
    for K in strike_prices:
        vols_SABR.append(SABR(F0, K, T, alpha, beta, rho, nu))
    return vols_SABR, alpha, rho, nu

vol_SABR_x1 = vol_SABR(SPX_df1_BS, F0_SPX1, strike_prices_x1_BS, r1, T1, "SPX", exdate1)[0]
vol_SABR_x2 = vol_SABR(SPX_df2_BS, F0_SPX2, strike_prices_x2_BS, r2, T2, "SPX", exdate2)[0]
vol_SABR_x3 = vol_SABR(SPX_df3_BS, F0_SPX3, strike_prices_x3_BS, r3, T3, "SPX", exdate3)[0]
vol_SABR_y1 = vol_SABR(SPY_df1_BS, F0_SPY1, strike_prices_y1_BS, r1, T1, "SPY", exdate1)[0]
vol_SABR_y2 = vol_SABR(SPY_df2_BS, F0_SPY2, strike_prices_y2_BS, r2, T2, "SPY", exdate2)[0]
vol_SABR_y3 = vol_SABR(SPY_df3_BS, F0_SPY3, strike_prices_y3_BS, r3, T3, "SPY", exdate3)[0]
```

Discuss how changing $\beta$ in the displaced-diffusion model and changing $\rho$ and $\nu$ in the SABR model affect the shape of the implied volatility smile.

**Answer:** In both the displaced-diffusion model and the SABR model, changes to specific parameters can significantly alter the implied volatility smile, which reflects how market volatility varies across different strike prices.

1. Displaced-diffusion<br>
In the displaced-diffusion model, changing $\beta$ affects the curvature and the skewness of the implied volatility smile:
- Lower $\beta$ (closer to 0): This results in a smile with more pronounced curvature, meaning that volatility for out-of-the-money (OTM) and in-the-money (ITM) options increases significantly, while at-the-money (ATM) options remain less volatile. A lower $\beta$ causes the model to behave more like a lognormal model, increasing the steepness of the volatility smile.
- Higher $\beta$ (closer to 1): As $\beta$ approaches 1, the model behaves more like a standard Black-Scholes model, where the volatility smile flattens. This results in less skewness and curvature, meaning the volatility across different strikes becomes more uniform.

2. SABR<br>
In the SABR model, $\rho$ (correlation between the asset price and volatility) and $\nu$ (volatility of volatility) play crucial roles in shaping the implied volatility smile:
- $\rho$: This parameter controls the skewness of the volatility smile.
>- $\rho<0$ leads to a downward sloping skew (right-skewed), where volatility decreases for higher strike prices and increases for lower strike prices. This is typical in markets where OTM puts are priced higher due to perceived higher risk (e.g., crash risk).
>- $\rho>0$ causes the skew to slope upward (left-skewed), where OTM calls become more expensive.

- $\nu$: This parameter controls the overall steepness and curvature of the volatility smile.
>- A higher $\nu$ increases the variability in volatility, making the smile more pronounced and increasing volatility for far OTM and ITM options.
>- A lower $\nu$ flattens the smile, leading to less variability in implied volatilities across different strikes.

Thus, adjusting these parameters allows both models to capture different features of the volatility smile observed in the market, with $\beta$ in the displaced-diffusion model focusing more on curvature, and $\rho$ and $\nu$ in the SABR model influencing skewness and the overall height of the smile.
